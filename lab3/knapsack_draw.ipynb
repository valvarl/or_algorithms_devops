```python
import json
import pandas as pd
import matplotlib.pyplot as plt

path = "../cmake-build-debug/lab3/test_results/knapsack/"
out_path = "test_results/knapsack/"
```

```python
data = pd.read_csv(path + "results.txt", header=None, index_col=0)
data.columns = [ "price", "weight", "time"]
data.index.name = "data"
```

```python
data
```

```python
with open(path + "values.txt") as inv:
    values = json.load(inv)

with open(path + "weights.txt") as inw:
    weights = json.load(inw)
```

```python
value_column = []
for i in range(7):
    plt.figure(i)
    plt.plot(values[i])
    plt.xlabel("generations")
    plt.ylabel("price")
    draw_path = f"value_track{i + 1}.png"
    value_column.append(draw_path)
    plt.savefig(out_path + draw_path)
```

```python
weight_column = []
for i in range(7):
    plt.figure(i)
    plt.plot(weights[i])
    plt.xlabel("generations")
    plt.ylabel("weight")
    draw_path = f"weight_track{i + 1}.png"
    weight_column.append(draw_path)
    plt.savefig(out_path + draw_path)
```

```python
pd.concat([pd.Series(value_column, name="price dependency", index=data.index), 
           pd.Series(weight_column, name="weight dependency", index=data.index),
          data],
          axis=1).to_csv("test_results/knapsack/output.txt")
```
