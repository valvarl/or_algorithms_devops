```python
import json
import pandas as pd
import matplotlib.pyplot as plt

path = "../cmake-build-debug/lab3/test_results/tsp/"
out_path = "test_results/tsp/"
```

```python
data = pd.read_csv(path + "results.txt", header=None, index_col=0)
data.columns = [ "cost", "time, s"]
data.index.name = "data"
```

```python
data
```

```python
with open(path + "costs.txt") as inc:
    costs = json.load(inc)
```

```python
cost_column = []
for i in range(6):
    plt.figure(i)
    plt.plot(costs[i])
    plt.xlabel("generations")
    plt.ylabel("cost")
    draw_path = f"{data.index[i]}.png"
    cost_column.append(draw_path)
    plt.savefig(out_path + draw_path)
```

```python
pd.concat([pd.Series(cost_column, name="cost dependency", index=data.index), 
          data], axis=1).to_csv("test_results/tsp/output.txt")
```
